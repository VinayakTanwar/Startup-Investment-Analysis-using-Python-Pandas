**Create a new column: 52 Week Range = 52 Week High - 52 Week Low**


```python
df.head()
```

```python
df['52 Week Range'] = df['52 Week High'] - df['52 Week Low']
df.head()
```

**Create a flag: Is_Undervalued = Price/Earnings < 15 (bool)**

```python
df['Undervalued'] = df['Price/Earnings'] < 15
df.head()
```

**Categorize Market Cap:

Small Cap (<10B)

Mid Cap (10Bâ€“100B)

Large Cap (>100B)**

```python
df['Small Cap'] = df['Market Cap'] < 10000000000
df['Mid Cap'] = (df['Market Cap'] >= 10000000000) & (df['Market Cap'] < 1000000000000)
df['Large Cap'] = df['Market Cap'] >= 1000000000000
df.head()
```

# **Data Visualization**

Bar plot of average Earnings/Share by Sector

Distribution plot of Price/Earnings

Heatmap of correlation between numerical columns

Scatter plot of Market Cap vs Price

Boxplot of Dividend Yield across sectors

```python
df.head()
df.groupby('Sector')['Earnings/Share'].mean().plot(kind = 'barh')
plt.title('Average Earnings/Share by Sector')
```

**Distribution plot of Price/Earnings**

```python
df.head()
sea.histplot(df['Price/Earnings'],  bins=30, kde=True)
```

**Heatmap of correlation between numerical columns**

```python
corr = df.corr(numeric_only=True)
sea.heatmap(corr, cmap='Blues',)
```

**Scatter plot of Market Cap vs Price**

```python
sea.scatterplot(x='Market Cap', y='Price', data=df, hue='Sector')
```

**Boxplot of Dividend Yield across sectors**

```python
sea.boxplot(x='Dividend Yield', y='Sector', data=df)
```

**Which sector has the highest average market cap?**

```python
df.head()
a = df.groupby('Sector')['Market Cap'].mean().sort_values(ascending = False)
sea.barplot(a)
```

**Which companies are undervalued based on Price/Earnings?**

```python
df.head()
df[df['Undervalued'] == True]['Name']
```

**Are higher dividends related to lower stock prices?**

```python
d = df[(df['Dividend Yield'] > 5) & (df['Price'] > 10)]
sea.barplot(x='Dividend Yield', y='Price', data=d)
```

**Top 5 Most Undervalued Stocks by P/E Ratio**

```python
df[df['Undervalued']].sort_values('Price/Earnings').head(5)
```

**Bar Plot: Number of Companies in Each Sector**

```python
sea.countplot(y='Sector', data=df)
```

**Pie Chart: Market Cap Distribution by Cap Category**

```python
df['Cap Type'] = df[['Small Cap', 'Mid Cap', 'Large Cap']].idxmax(axis=1)
df['Cap Type'].value_counts().plot.pie(autopct='%1.1f%%')
```
